1. What is the problem with simply labeling customers are "churned" or "not churned" and using standard supervised learning techniques?

The issue is that ‚Äúchurned"/"not churned‚Äù throws away the timing information. If we just mark people as ‚Äúchurned‚Äù or ‚Äúnot churned,‚Äù we‚Äôre ignoring when things happened. Lots of customers in the dataset haven‚Äôt left yet, so we don‚Äôt actually know their true churn time, we only know that they‚Äôve stayed at least as long as their current tenure. A regular supervised model doesn‚Äôt understand this and treats those cases as if we fully observed the outcome, which isn‚Äôt true. This is why we end up with biased predictions, usually underestimating how long people stay.

2. Consider customer A who just joined last week vs. customer B who has been with the service for a year. Who do you expect will leave the service first: probably customer A, probably customer B, or we don't have enough information to answer? Briefly explain your answer. 

Customer A has only been here a week, while B has already made it through a full year. From the survival plots in the lecture, people who last longer tend to have a lower hazard going forward. So I‚Äôd expect Customer A to be more likely to leave first. The key idea is that surviving longer already gives information about their risk of churn.

3. If a customer's survival function is almost flat during a certain period, how do we interpret that?

If the survival curve barely moves over some period, it means almost nobody is churning in that window. In other words, the hazard is very low there, customers who make it to that part of the timeline tend to stick around through it.

```python
# BLOG DATASET SETUP - FROM HW 8

from hashlib import sha1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, OneHotEncoder

from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import r2_score

df = pd.read_csv("avocado.csv", parse_dates=["Date"], index_col=0)

split_date = '20170925'
df_train = df[df["Date"] <= split_date]
df_test  = df[df["Date"] >  split_date]

import pandas as pd


def create_lag_feature(
    df: pd.DataFrame,
    orig_feature: str,
    lag: int,
    groupby: list[str],
    new_feature_name: str | None = None,
    clip: bool = False,
) -> pd.DataFrame:
    """
    Create a lagged (or ahead) version of a feature, optionally per group.

    Assumes df is already sorted by time within each group and has unique indices.

    Parameters
    ----------
    df : pd.DataFrame
        The dataset.
    orig_feature : str
        Name of the column to lag.
    lag : int
        The lag:
          - negative ‚Üí values from the past (t-1, t-2, ...)
          - positive ‚Üí values from the future (t+1, t+2, ...)
    groupby : list of str
        Column(s) to group by if df contains multiple time series.
    new_feature_name : str, optional
        Name of the new column. If None, a name is generated automatically.
    clip : bool, default False
        If True, drop rows where the new feature is NaN.

    Returns
    -------
    pd.DataFrame
        A new dataframe with the additional column added.
    """
    if lag == 0:
        raise ValueError("lag cannot be 0 (no shift). Use the original feature instead.")

    # Default name if not provided
    if new_feature_name is None:
        if lag < 0:
            new_feature_name = f"{orig_feature}_lag{abs(lag)}"
        else:
            new_feature_name = f"{orig_feature}_ahead{lag}"

    df = df.copy()

    # Map your convention (negative=past, positive=future) to pandas shift
    # pandas: shift(+k) ‚Üí past, shift(-k) ‚Üí future
    periods = abs(lag) if lag < 0 else -lag

    df[new_feature_name] = (
        df.groupby(groupby, sort=False)[orig_feature]
          .shift(periods)
    )

    if clip:
        df = df.dropna(subset=[new_feature_name])

    return df


df_sort = df.sort_values(by=["region", "type", "Date"]).reset_index(drop=True)
df_hastarget = create_lag_feature(df_sort, "AveragePrice", +1, ["region", "type"], "AveragePriceNextWeek", clip=True)

df_train = df_hastarget[df_hastarget["Date"] <= split_date]
df_test  = df_hastarget[df_hastarget["Date"] >  split_date]
```

# BLOG: HW8 Question 1.5: Forecasting average avocado price

For this blog post, I revisited question 1.5 from HW5 where we tried to forecast the average price of avocados one week ahead. # Description of the problem/decision.

The main question was "can we predict next week‚Äôs avocado price using historical data?" Because the target is just the price shifted forward by one week, the problem falls between standard supervised learning and time-series forecasting. We also have different regions and two avocado types (conventional vs organic) - which introduces extra variability. Our goal, according to the question specification, is to aim for an R^2 score of >= 0.79 on our test split.

# Description of the dataset (the raw data and/or some EDA).

The avocado.csv dataset contains weekly prices throughout various regions. I mainly used:
- AveragePrice
- Date
- Region
- Type

I then created the target column named AveragePriceNextWeek by shifting the price values (lag). One challenge regarding this avocado dataset is that the date column cannot be utilized as a timestamp. The model won't understand weekly or monthly patterns unles we engineer them (by encoding them). This was key.

Before any modelling, I took a quick look at the relaitonship between average price and month (EDA), since avocado demand/price is seasonal. The visualization below in Figure 1 helped confirm that prices vary per month.

```python
# FROM HW 8
df_sort = df.sort_values(by=["region", "type", "Date"]).reset_index(drop=True)
df_hastarget = create_lag_feature(df_sort, "AveragePrice", +1,
                                 ["region", "type"], "AveragePriceNextWeek",
                                 clip=True)


df_hastarget["year2"]  = df_hastarget["Date"].dt.year
df_hastarget["month2"] = df_hastarget["Date"].dt.month
df_hastarget["week2"]  = df_hastarget["Date"].dt.isocalendar().week.astype(int)

df_month = df_hastarget.groupby("month2")["AveragePrice"].mean()
df_month.plot(kind="line")
plt.title("Figure 1: Average Avocado Price by Month")
plt.xlabel("Month")
plt.ylabel("Average Price")
plt.show()
```

As shown in Figure 1, some months consistently have a higher average avocado price, which justified splitting the date column into separate numeric components.


# Description of the model/Modelling Approach
To give the model more structure, I broke the Date column into:
- year2
- month2
- week2 

NUmeric features were passed through with no change whereas Region and Type went through one-hot encoding using a column transformer.

I tested a few ML models like linear regression, ridge regression, and random forest - but the one that performed best was GradientBoostingRegressor. 

To avoid overfitting, I did not finetune it much.

Before trusting the model, I visualized how different regions behaved on average (region appeared to be one of the strongest indicator features). The bar chart below in Figure 2 made it clear regions differ quite a lot.

```python
# FROM HW 8
df_region = (
    df_hastarget.groupby("region")["AveragePrice"]
    .mean()
    .sort_values()
)

plt.figure(figsize=(12,5))
df_region.plot(kind="bar")

plt.title("Figure 2: Average Price by Region")
plt.ylabel("Average Price")
plt.xlabel("Region")

plt.show()

```

Figure 2 shows that some areas consistently have higher or lower prices, meaning region is an important signal the model needs to aknowledge.

# Description your results, both quantitatively and qualitatively. Make sure to refer to the original problem/decision.

After training the GradientBoostingRegressor model, the R^2 scores were:

Train Split R¬≤ = 0.858
Test Split R¬≤ = 0.797

This clears the benchmark without having the model experience overfitting. This means we can realistically expect our model to generalize on deployment data!

To see how the predictions/results lined up with the actual prices/target, I plotted the model's predicted values (predicitons) across the testing period timerange. The visualiztion below in Figure 3 was useful for checking that the model wasn't producing unrealistic price predicitons due to not being generalizable.


```python
#REFERENCES: IDEA for visualization given by ChatGPT (however, implementation was not provided by ChatGPT). I asked and confirmed that this was allowed. 
# Implementation was aided by online resources on data visualization in Python
# Much of the implementation was my work from HW 8.


#From HW 8 
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import GradientBoostingRegressor

# Date encodings
df_has_target = df_hastarget.copy()
df_has_target["year2"] = df_has_target["Date"].dt.year
df_has_target["month2"] = df_has_target["Date"].dt.month
df_has_target["week2"] = df_has_target["Date"].dt.isocalendar().week.astype(int)

# Re-split
df_train = df_has_target[df_has_target["Date"] <= split_date]
df_test  = df_has_target[df_has_target["Date"] >  split_date]

# Features
X_train = df_train[["AveragePrice", "year2", "month2", "week2", "region", "type"]]
y_train = df_train["AveragePriceNextWeek"]

X_test  = df_test[["AveragePrice", "year2", "month2", "week2", "region", "type"]]
y_test  = df_test["AveragePriceNextWeek"]

from sklearn.compose import make_column_transformer

preprocess = make_column_transformer(
    (OneHotEncoder(handle_unknown="ignore"), ["region", "type"]),
    ("passthrough", ["AveragePrice", "year2", "month2", "week2"])
)


model = make_pipeline(
    preprocess,
    GradientBoostingRegressor(random_state=123)
)
model.fit(X_train, y_train)

# Aggregate actual values by date
df_daily_actual = (
    df_test.groupby("Date")["AveragePriceNextWeek"]
    .mean()
    .reset_index()
)

# Aggregate predictions by date (match index first)
pred = model.predict(X_test)
pred_series = pd.Series(pred, index=df_test.index)

df_daily_pred = (
    pred_series.groupby(df_test["Date"])
    .mean()
    .reset_index(name="PredictedPrice")
)

# Merge two series to align dates 1-to-1
df_plot = df_daily_actual.merge(
    df_daily_pred,
    on="Date",
    how="inner"
)

plt.figure(figsize=(14,6))

plt.plot(
    df_plot["Date"],
    df_plot["AveragePriceNextWeek"],
    label="Actual Price",
    linewidth=2
)

plt.plot(
    df_plot["Date"],
    df_plot["PredictedPrice"],
    label="Predicted Price",
    linewidth=2
)

plt.title("Figure 3: Actual vs Predicted Avocado Prices (Test Set)")
plt.xlabel("Date")
plt.ylabel("Average Price")
plt.legend()
plt.show()
```

As seen in Figure 3, the model accurately predicts the avocado price movement fairly well. This reasssures us that the model learned meaningful patterns rather than memorizing noise from our train split or bias from underfitting!

Predicitons seem to be driven by 3 things:
1. Seasonality (captured by year/month/week breakdowns)
2. Region (captured by one hot encoding)
3. Gradient Boosting Model being able to conform to non-linear data

This did not feel like "fishing" for a manually created high test score - the improvements over the baseline model were consistent with the data.

# A section on caveats, describing at least 3 reasons why your results might be incorrect, misleading, overconfident, or otherwise problematic. Make reference to your specific dataset, model, approach, etc. 
Even though the model performed decently, there are a few important limitations:

1. Only one single time-based split

With time-series data we cannot shuffle, so the results of the model depend heavily on the exact split. More splits (for xample, TimeSeriesSplit) would give us more reliable and stable estimates.

2. Simple date encoding

Year, month, and week capture basic patterns. However, more nuanced patterns like supply chain disriptions, holidays, and locational seasonality) are ignored within our current model.

3. Tree-based models are likely to overfit/overtrain on subtle patterns reducing gernealizability

Even without hyperparameter tuning, gradient boosting can still latch onto noise. Especially when there are dozens of one-hot encoded regions in the dataset.

Because of these caveat issues, I personally wouldn‚Äôt treat the R^2 score as a definitive measure of the real-world deployment data. But for the purposes of CPSC 330, and given the limited number of allowed test split runs, this approach feels satisfactory.
